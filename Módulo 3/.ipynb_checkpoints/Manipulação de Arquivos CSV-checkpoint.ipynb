```python
import csv
```

```python
with open('Brasil_covid.csv', 'r', encoding='utf-8') as arquivo_csv:
    leitor = csv.reader(arquivo_csv)
    next(leitor) #Pula uma das linhas
    for linha in leitor:
        if int(linha[3]) > 1:
            print(linha)
```

```python
with open('Brasil_covid.csv', 'r', encoding = 'utf-8') as arquivo_csv:
    linhas = arquivo_csv.read()
    linhas = linhas.split('\n')
    for linha in linhas:
        linha = linha.split(',')
        print(linha)
```

```python
with open('user.csv', 'w', encoding = 'utf-8', newline='') as arquivo_users:
    escritor = csv.writer(arquivo_users)
    escritor.writerow(['nome','sobrenome', 'email', 'genero'])
    escritor.writerow(['Yago', 'Henrique', 'yago@email.com', 'masculino'])
```

```python
header = ['nome', 'sobrenome']
dados = []
opt = int(input('O que deseja fazer?\n [1] - Cadastrar\n [0] - Sair\n'))
while opt != 0:
    nome = input("Digite o nome para ser cadastrado:\n")
    sobrenome = input('Digite o sobrenome para ser cadastrado:\n')
    dados.append([nome, sobrenome])
    opt = int(input('O que deseja fazer?\n [1] - Cadastrar\n [0] - Sair\n'))

with open('users.csv', 'w', encoding='utf-8', newline='') as arquivo_csv:
    writer = csv.writer(arquivo_csv)
    writer.writerow(header)
    writer.writerows(dados)
    
with open('users.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    for linha in csv_reader:
        print(linha)
```
